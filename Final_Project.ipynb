There is no information on total pit-stops so we will have to extract it from this table

```python
pits.info()
```

```python
totalPits = pd.DataFrame(columns = ['raceId', 'driverId', 'totalPits'])
```

```python
totalPits
```

```python
for race in pits['raceId'].unique():
    for driver in pits[pits['raceId'] == race]['driverId'].unique():
        dtotalPits = max(pits[(pits['raceId'] == race) & (pits['driverId'] == driver)]['stop'])
        totalPits.loc[len(totalPits.index)] = {'raceId' : race, 'driverId' : driver, "totalPits" : dtotalPits}
totalPits
```

Now we will validate the target variable

```python
import numpy as np
np.unique(dataTrim['points'])
```

according to F1 rules over the sport's history, the maximum points for one race is 26, 25 for first place plus 1 for the fastest lap, so I will have to explore these outliers

The winner receives 25 points, the second-place finisher 18 points, with 15, 12, 10, 8, 6, 4, 2 and 1 points for positions 3 through 10, respectively. One additional point is awarded to the driver and team with the fastest lap of the race, but only if this driver finishes in the top 10 positions.

```python
dataTrim[dataTrim['points'] > 26]
```

Again, all these outliers correspond to a single race. Exploring this race's history, I found that there was a double points rule at play when it was ran. This was done to keep the competition alive. This resulted in the odd 50 point race by Hamilton. To correct this, I will divide all scores from this race by 2.

```python
dataTrim.loc[dataTrim['raceId'] == 918]
```

```python
dataTrim.loc[dataTrim['raceId'] == 918, 'points'] = dataTrim.loc[dataTrim['raceId'] == 918, 'points']/2
np.unique(dataTrim['points'])
```

Binning the target variable to make predictions easier and more accurate

```python
choices = [
    ((dataTrim['points'] > 0), 'Scoring'),
    ((dataTrim['points'] < 1), 'Not-scoring')
]

dataTrim['pointsBin'] = np.select(*zip(*choices), default='Other')
dataTrim
```

Modeling

```python
x = dataTrim[['driverRef', 'circuitRef',  'constructorRef', 'totalPits', 'year']]
y = dataTrim['pointsBin']
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
xEnc = enc.fit_transform(x[['driverRef', 'constructorRef', 'circuitRef']]).toarray()
xEnc = pd.DataFrame(xEnc, columns=enc.get_feature_names_out(x[['driverRef', 'constructorRef', 'circuitRef']].columns))
xEnc
```

```python
x = x.merge(xEnc, right_index = True, left_index = True)
x
```

```python
x = x.drop(columns = ['driverRef', 'constructorRef', 'circuitRef'])
x
```

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(xTrain, yTrain)
preds = rf.predict(xTest)
```

```python
predsT = rf.predict(xTrain)
```

```python
from sklearn import metrics
metrics.accuracy_score(yTrain, predsT)
```

```python
from sklearn import metrics
metrics.accuracy_score(yTest, preds)
```

```python
metrics.confusion_matrix(yTest, preds)
```
